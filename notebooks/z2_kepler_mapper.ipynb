## L-infinity centrality

```python
M = horse_df.as_matrix()
D = distance_matrix(M, M)
```

```python
L_inf_lens = D.max(axis=0)

mapper = km.KeplerMapper(verbose=1)

horse_L_inf = mapper.map(L_inf_lens,                          
                         M,          
                         clusterer=DBSCAN(eps=0.05),
                         coverer=Cover(nr_cubes=20, overlap_perc=0.3))

mapper.visualize(horse_L_inf,   
                 color_function=L_inf_lens,
                 path_html="horse.html",
                 title="horse, L_inf centrality lens")
```
