```python
import pandas as pd

from umap import UMAP
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
```

```python
digits = load_digits()
```

```python
digits.data.shape
```

```python
len(digits.data)
```

```python
UMAP(n_neighbors=5, min_dist=0.3, metric='cosine').fit_transform
```

```python
UMAP(n_neighbors=5, min_dist=0.3, metric='cosine').fit_transform
```

```python
%%time
embedding = UMAP(n_neighbors=5,
                 min_dist=0.3,
                 metric='cosine').fit_transform(digits.data)
```

```python
embedding_df = pd.DataFrame(embedding, columns=['umap_x', 'umap_y'])
```

```python
embedding_df.plot.scatter(figsize=(8, 8), x='umap_x', y='umap_y', c=digits.target, cmap='Set1', alpha=0.5)
plt.show()
```

```python
%%time
embedding = UMAP(n_neighbors=5,
                 min_dist=0.3,
                 metric='euclidean').fit_transform(digits.data)
```

```python
embedding_df = pd.DataFrame(embedding, columns=['umap_x', 'umap_y'])
```

```python
embedding_df.plot.scatter(figsize=(8, 8), x='umap_x', y='umap_y', c=digits.target, cmap='Set1', alpha=0.5)
plt.show()
```
